```python
import pygame, math, random
from pygame import key
screen = pygame.display.set_mode([400,400])
clock = pygame.time.Clock()
pygame.init()
# Classes(major): Debuff, Tank, Assasssin, Support, Melee, Ranger
class Unit(pygame.sprite.Sprite):
    def __init__(self, name, x, y, major,team):
        # Call the parent's constructor
        super().__init__()

        # Set height, width
        self.image = pygame.Surface([10,10])
        self.image.fill((255,255,255))
        # Make our top-left corner the passed-in location.
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y
        self.name = name
        self.team = team
        if major == 'D':
            setting = [50,50,5,50]
        if major == 'T':
            setting = [150,150,15,30]
        if major == 'A':
            setting = [50,50,25,20]
        if major == 'S':
            setting = [75,75,0,200]
        if major == 'M':
            setting = [100,100,10,40]
        if major == 'R':
            self.hp_mx = 75
            self.hp = 75
            self.dmg = 15  
            self.range = 125
            setting = [75,75,15,125]
        self.hp_max = setting[0]
        self.hp = setting[1]
        self.dmg = setting[2]
        self.range = setting[3]
    def take_dmg(self,dmg_taken):
        self.image.fill((255,0,0))
        self.hp -= dmg_taken
        print(self, "lost", dmg_taken, "health")
    def healthbar(self,hp,hp_mx):
        pass
# pygame.draw.rect(screen,((255,255,255)),50,50)
# do white hp max
# do green to over ride with hp
    def movement(self,direction):
        if direction == 1:
            self.rect.y -= 5
        if direction == 2:
            self.rect.x += 5
        if direction == 3:
            self.rect.y += 5
        if direction == 4:
            self.rect.x -=5
                
steve = Unit('Steve',200,100,'M','A')
alex = Unit('Alex',100,100,'R','A')
bob = Unit('Bob',100,200,'R','B')
joe = Unit('Joe',200,200,'M','B')
TeamA = pygame.sprite.Group(steve,alex)
TeamB = pygame.sprite.Group(bob,joe)
TeamAlist = [steve,alex]
TeamBlist = [bob,joe]
TeamAandBlist = [steve,alex,bob,joe]
TeamAview = {} #updated A not B
TeamBview = {} #updated B not A
new_map = {} #updated A and B
#creating initial views and map
for sprite in TeamAandBlist:
    TeamAview[sprite] = [sprite.rect.x,sprite.rect.y,sprite.name]
    TeamBview[sprite] = [sprite.rect.x,sprite.rect.y,sprite.name]
#end of intial views and map
#set variables
running = True
moving = True
sight = False
#end of setting variables
#movements and ranges
while running:
    i=0
    screen.fill([0,0,0])
    for unit in TeamAandBlist:
        #set screen for A or B
        if TeamAandBlist[i].team == 'A':
            for k in TeamAandBlist:
                k.rect.x = TeamAview[k][0]
                k.rect.y = TeamAview[k][1]
            TeamA.draw(screen)
            TeamB.draw(screen)    
            pygame.display.flip()
            
        if TeamAandBlist[i].team == 'B':
            for k in TeamAandBlist:
                k.rect.x = TeamBview[k][0]
                k.rect.y = TeamBview[k][1]
            TeamA.draw(screen)
            TeamB.draw(screen)    
            pygame.display.flip()
        #end of setting screen for A or B
        #check if limbo
        limbo = False
        if TeamAandBlist[i].team != TeamAandBlist[i-1].team: #if different teams
            limbo = True
            print('SWITCH')
            print('SPACE TO CONTINUE')  
        while limbo == True:
            print('kys')
            pygame.draw.rect(screen,(255,255,255),(200,200,200,200))
            for event in pygame.event.get():
                pygame.draw.rect(screen,(255,255,255),(200,200,200,200)) 
                if event.type == pygame.QUIT:
                    pygame.quit()
                    running = False
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_ESCAPE:
                        pygame.quit()
                        running = False
                    if event.key == pygame.K_SPACE:
                        limbo = False
                
        #end of limbo                
        #check for key inputs
        movements = 0
        while movements < 10:
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    pygame.quit()
                    running = False
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_ESCAPE:
                        pygame.quit()
                        running = False
                    if event.key == pygame.K_q:
                        movements = 10
                    if event.key == pygame.K_m:
                        movements -= 1
                        if sight == True:
                            sight = False
                        else:
                            sight = True
                    if event.key == pygame.K_p:
                        print(new_map)                        
                    else:
                        print(movements + 1)
                        movements += 1
                        if event.key == pygame.K_UP:
                            unit.movement(1)
                        if event.key == pygame.K_RIGHT:
                            unit.movement(2)
                        if event.key == pygame.K_DOWN:
                            unit.movement(3)
                        if event.key == pygame.K_LEFT:
                            unit.movement(4)
                screen.fill([0,0,0]) 
        #end of key inputs
        #check for sight
            color = random.randint(10,200)        
            if sight == True:
                for sprite in TeamA:
                    pygame.draw.circle(screen,(color,0,0),(sprite.rect.x+5,sprite.rect.y+5),sprite.range)
                for sprite in TeamB:
                    pygame.draw.circle(screen,(0,0,color),(sprite.rect.x+5,sprite.rect.y+5),sprite.range)
            #end of sight

            TeamA.draw(screen)
            TeamB.draw(screen)    
            pygame.display.flip()
            

        if unit.team == 'A':
            TeamAview[unit] = (unit.rect.x, unit.rect.y, unit.name)
        if unit.team == 'B':
            TeamBview[unit] = (unit.rect.x, unit.rect.y, unit.name)
        new_map[unit] = (unit.rect.x, unit.rect.y, unit.name)
        print('This sprite is done moving')
        i+=1
        #end of for each unit
    print('All movements are done')
    print(TeamAview,'\n',TeamBview,'\n',new_map)
    #ALL MOVEMENTS ARE DONE
    
    #both players done moving
    #checking if in ranges
    for unita in TeamAlist:
        unita.radius = unita.range
        for unitb in TeamBlist:
            if pygame.sprite.collide_circle(unita,unitb):
                print(unitb.name, ' is in the range of ',unita.name)
        unita.radius = 5
    for unitb in TeamBlist:
        unitb.radius = unitb.range
        for unita in TeamAlist:
            if pygame.sprite.collide_circle(unitb,unita):
                print(unita.name, ' is in the range of ',unitb.name)
        unitb.radius = 5

    clock.tick(5)
pygame.quit()

```
